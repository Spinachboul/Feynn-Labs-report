```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the dataset
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
```

```python
df.head()
```

```python
# Create a bar chart showing the number of vehicles in each state
state_counts = df['State'].value_counts()
plt.figure(figsize=(15,12))
plt.bar(state_counts.index, state_counts.values)
plt.xlabel('State')
plt.ylabel('Number of Electric Vehicles')
plt.show()
```

```python
# Create a map showing the distribution of electric vehicles by postal code
postal_code_counts = df['Postal Code'].value_counts()
plt.figure(figsize=(10,10))
sns.heatmap(postal_code_counts.to_frame().T, annot=True, cmap='Blues')
plt.xlabel('Postal Code')
plt.ylabel('Number of Electric Vehicles')
plt.show()
```

```python
# Create a line chart showing the number of electric vehicles registered by model year
model_year_counts = df['Model Year'].value_counts()
plt.plot(model_year_counts.index, model_year_counts.values)
plt.xlabel('Model Year')
plt.ylabel('Number of Electric Vehicles registered')
plt.show()
```

```python
# Create a pie chart showing the most popular electric vehicle makes
make_counts = df['Make'].value_counts()
plt.figure(figsize=(20,30))
plt.pie(make_counts.values, labels = make_counts.index, autopct="%1.1f%%")
plt.show()
```

```python
# Create a bar chart showing the most popular electric vehicle models
model_counts = df['Model'].value_counts()
plt.figure(figsize=(80,10))
plt.bar(model_counts.index, model_counts.values)
plt.xlabel("Model")
plt.ylabel("Number of Electric Vehicles")
plt.show()
```

```python
# Create a bar chart that shows the number of electric vehicles that are eligible for CAFV incentives
cafv_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
plt.figure(figsize=(10,20))
plt.bar(['Not eligible due to low battery range','Clean Alternative Fuel Vehicle Eligible'], [cafv_counts[1], cafv_counts[0]])
plt.xlabel('CAFV Eligibility')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Create a barplot that shows the electric vehicle ranges
range_counts = df['Electric Range'].value_counts()
plt.bar(range_counts.index, range_counts.values)
plt.xlabel('Electric Range')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Create a scatter plot showing the relationship between the base MSRP and electric range
plt.scatter(df['Base MSRP'], df['Electric Range'])
plt.xlabel('Base MSRP')
plt.ylabel('Electric Range')
plt.show()
```

```python
# Create a map showing the distribution of vehicles by legislative district
legislative_district_counts = df["Legislative District"].value_counts()
plt.figure(figsize=(50, 10))
sns.heatmap(legislative_district_counts.to_frame().T, annot=True, cmap="Blues")
plt.xlabel("Legislative District")
plt.ylabel("Number of Electric Vehicles")
plt.show()
```
